```python
from abc import ABC, abstractmethod
```

```python
class employee(ABC):
    @abstractmethod
    def calculate_salary(self):
        pass
    @abstractmethod
    def details(self):
        pass
```

```python
class employees(employee):
    def __init__(self):
        self.name=input('enter name')
        self.salary=int(input('enter salary'))
        self.pf=self.salary*0.20
        self.calculate_salary()
    def calculate_salary(self):
        self.net_salary=self.salary+self.pf
        print(self.net_salary)
    def details(self):
        print( f"name of employee is {self.name} and its salary is{self.net_salary}")
    
```

```python
user1=employees()
```

```python
user1.details()
```

```python
a=int(input('num1'))
b=int(input('num2'))
print(a/b)
```

```python
try:
    a=int(input('num1'))
    b=int(input('num2'))
    print(a/b)
except(ZeroDivisionError):
    print('number can not be divided by zero')
except(ValueError):
    print('enter number only')
finally:
    print('program end successfully')
```

```python
try:
    a=int(input('num1'))
except(ValueError):
    print('taking default numbers')
    a=0
try:
    b=int(input('num2'))
except(ValueError):
    print('taking default number')
    b=0
try:
    print('div',a/b)
except(ZeroDivisionError):
    print('can not be dicided by zero')
finally:
    print('mul',a*b)
    print('sum',a+b)
```
