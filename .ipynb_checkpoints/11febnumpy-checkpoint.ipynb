```python
nums3=[]
for i in lst3:
    try:
        val=float(i)
        if not (val==float('1e3') or val==float('inf') or val==float('-inf') or val!=val):
            nums3.append(val)
    except:
        pass
```

```python
expenses=np.array([
    [200,340,230,150,456],
     [100,345,122,600,233],
     [350,450,233,122,323]
])
```

```python
print('total expense:',np.sum(expenses))
print('category wise sum:',np.sum(expenses,axis=1))
print('day wise total:',np.sum(expenses,axis=0))
print('highest expense day:',np.max(np.sum(expenses,axis=0)))
```

```python
marks= np.array([88,33,44,87,54,32,44,57,67])
```

```python
grades = np.where(marks >= 75, "A",
         np.where(marks >= 55, "B",
         np.where(marks >= 45, "C", "fail")))
```

```python
print(grades)
```
