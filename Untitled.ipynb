## comparison operator

```python
a = 10
b = 5

print(a == b)   # Equal to → False
print(a != b)   # Not equal to → True
print(a > b)    # Greater than → True
print(a < b)    # Less than → False
print(a >= b)   # Greater than or equal to → True
print(a <= b)   # Less than or equal to → False
```

## assignment op

```python
a = 10
a += 5
print(a)     # 15

a -= 3
print(a)     # 12

a *= 2
print(a)     # 24

a /= 4
print(a)     # 6.0

a //= 2
print(a)     # 3.0

a %= 3
print(a)     # 0.0

a **= 2
print(a)     # 0.0
```

## logical op

```python
a = 10
b = 5

# AND operator
print(a > 5 and b < 10)    # True

# OR operator
print(a < 5 or b < 10)     # True

# NOT operator
print(not(a > 5))          # False
```

```python
age = int(input("Enter age: "))

print(age >= 18 and age <= 60)

```

## bitwise op

```python
a = 10   # 1010
b = 4    # 0100

print(a & b)   # AND  -> 0
print(a | b)   # OR   -> 14
print(a ^ b)   # XOR  -> 14
print(~a)      # NOT  -> -11
print(a << 1)  # Left Shift  -> 20
print(a >> 1)  # Right Shift -> 5
```

## members op

```python
lst = [1, 2, 3, 4, 5]
name = "harsh"
s = {10, 20, 30}

print(3 in lst)          # True
print(6 in lst)          # False

print("h" in name)       # True
print("z" not in name)   # True

print(20 in s)           # True
print(40 not in s)       # True
```

## identity op

```python
a = 10
b = 10
c = [1, 2, 3]
d = [1, 2, 3]

print(a is b)        # True
print(a is not b)    # False

print(c is d)        # False
print(c == d)        # True
```

#### ONE SHOT EXAMPLE

```python
# Arithmetic Operators
a = 10
b = 3
print("Arithmetic:")
print(a + b)    # Addition
print(a - b)    # Subtraction
print(a * b)    # Multiplication
print(a / b)    # Division
print(a // b)   # Floor Division
print(a % b)    # Modulus
print(a ** b)   # Power


# Assignment Operators
print("\nAssignment:")
x = 5
x += 2
print(x)
x *= 3
print(x)


# Comparison Operators
print("\nComparison:")
print(a > b)
print(a == b)
print(a != b)


# Logical Operators
print("\nLogical:")
print(a > 5 and b < 5)
print(a < 5 or b < 5)
print(not(a < 5))


# Bitwise Operators
print("\nBitwise:")
print(a & b)
print(a | b)
print(a ^ b)
print(a << 1)
print(a >> 1)


# Membership Operators
print("\nMembership:")
lst = [1, 2, 3, 10]
print(10 in lst)
print(5 not in lst)


# Identity Operators
print("\nIdentity:")
m = [1, 2, 3]
n = [1, 2, 3]
print(m is n)
print(m == n)
```

```python
choice=int(input())
match choice:
     case 1:
         print('one')
     case 2:
         print('two')
     case 3:
         print('three')
     case _:
         print('invalid')
```

```python
aa=int(input('num1'))
bb=int(input('num2'))
op=input('operations')
match op:
     case '+':
         print(aa+bb)
     case '-':
         print(aa-bb)
     case '*':
         print(aa * bb)
     case _:
         print('invalid op')
```
