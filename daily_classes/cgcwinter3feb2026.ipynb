```python
ops=['+', '-', '*']
```

```python
print('calculator')
choice=input('enter choice')
num1=int(input('num1'))
num2=int(input('num2'))
if choice == 'add':
    add(num1, num2)
elif choice == 'sub':
    sub(num1, num2)
elif choice == 'mul':
    mul(num1, num2)
else:
    print('wrong choice')
```

```python
print('calculator')
print(f"+ for addition")
print(f"- for subtraction")
print(f"* for multiplication")
num1=int(input('num1'))
num2=int(input('num2'))
choice=input('enter your op +, - or *')
if choice in ops:
    if choice == '+':
        add(num1, num2)
    elif choice == '-':
        sub(num1, num2)
    elif choice == '*':
        mul(num1, num2)
else:
    print('enter correct choice')
```

```python
def new_calc():
    print('calculator')
    print(f"+ for addition")
    print(f"- for subtraction")
    print(f"* for multiplication")
    num1=int(input('num1'))
    num2=int(input('num2'))
    choice=input('enter your op +, - or *')
    if choice in ops:
        if choice == '+':
            add(num1, num2)
        elif choice == '-':
            sub(num1, num2)
        elif choice == '*':
            mul(num1, num2)
    else:
        print('enter correct choice')
```

```python
passwords=[]
users=[]
```

```python
username=input('enter user name')
password=input('enter password')
if len(password)<6:
    print('password too short')
elif ' ' in password:
    print('no space allowed')
elif not any (i in '!@#$%^&*()_+' for i in password ):
    print('password must have special characters')
elif not any (char.isupper() for char in password):
    print('password must have upper chars')
elif not any (char.islower() for char in password):
    print('password must have small letter')
elif not any (char.isdigit() for char in password):
    print('password must have numbers')
else:
    print('password is strong')
    passwords.append(password)
    users.append(username)
```

```python
user1=input('enter username')
if user1 in users:
    idx=users.index(user1)
    password1=input('enter password')
    if password1==passwords[idx]:
        new_calc()
    else:
        print('invalid password')
else:
    print('invalid user')
```

```python
def secure_calc():
    user1=input('enter username')
    if user1 in users:
        idx=users.index(user1)
        password1=input('enter password')
        if password1==passwords[idx]:
            new_calc()
        else:
            print('invalid password')
    else:
        print('invalid user')
```

```python
def check():
    username=input('enter user name')
    password=input('enter password')
    if len(password)<6:
        print('password too short')
    elif ' ' in password:
        print('no space allowed')
    elif not any (i in '!@#$%^&*()_+' for i in password ):
        print('password must have special characters')
    elif not any (char.isupper() for char in password):
        print('password must have upper chars')
    elif not any (char.islower() for char in password):
        print('password must have small letter')
    elif not any (char.isdigit() for char in password):
        print('password must have numbers')
    else:
        print('password is strong')
        passwords.append(password)
        users.append(username)
```

```python
operation=input('enter choice')
if operation=='password':
    check()
elif operation == 'calc':
    secure_calc()
else:
    print('wrong choice')
```

```python
def final():
    operation=input('enter choice')
    if operation=='password':
        check()
    elif operation == 'calc':
        secure_calc()
    else:
        print('wrong choice')
```
